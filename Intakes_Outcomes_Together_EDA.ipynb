```python
import pandas as pd
```

```python
intakes = pd.read_csv('Cleaned_Intakes.csv')
outcomes = pd.read_csv('Cleaned_Outcomes.csv')
```

```python
merged_outer = intakes.merge(outcomes,how='outer',on='Animal ID',suffixes=('_intakes','_outcomes'))
merged_outer.head()
```

```python
merged_outer.info()
```

```python
merged_inner = intakes.merge(outcomes,how='inner',on='Animal ID')
merged_inner.head()
```

```python
merged_inner.info()
```

```python
print("\n")
print("Total cases of intakes and outcomes "+ str(intakes.shape[0]+ outcomes.shape[0]))
print("\n")
print("The number of casee both intake and outcome happened " + str(merged_inner.shape[0]))
print("\n")
print("The number of intake cases without outcomes" + str(intakes.shape[0] - merged_inner.shape[0] ))
print("\n")
print("The number of outcome cases without intake" + str(outcomes.shape[0] - merged_inner.shape[0]))
print("\n")
```
